## 1. Importación de Librerías

```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from scipy.optimize import linprog
import time
from collections import deque

plt.style.use('seaborn-v0_8-whitegrid')
np.random.seed(42)
```

## 2. Lectura de la Instancia DIMACS

La instancia C125.9 es un grafo aleatorio con 125 nodos y probabilidad de arista 0.9.  
Formato DIMACS: líneas `p col n m` (problema) y `e u v` (aristas).

```python
def parse_dimacs(filename):
    """Parsea un archivo DIMACS .clq y retorna un grafo NetworkX."""
    G = nx.Graph()
    with open(filename, 'r') as f:
        for line in f:
            if line.startswith('p'):
                parts = line.split()
                n_vertices = int(parts[2])
                G.add_nodes_from(range(1, n_vertices + 1))
            elif line.startswith('e'):
                parts = line.split()
                u, v = int(parts[1]), int(parts[2])
                G.add_edge(u, v)
    return G

G = parse_dimacs('C125.9.clq')
n = G.number_of_nodes()
m = G.number_of_edges()
density = 2 * m / (n * (n - 1))

print(f"Instancia: C125.9")
print(f"Nodos: {n}")
print(f"Aristas: {m}")
print(f"Densidad: {density:.4f}")
print(f"Aristas complemento (restricciones): {n*(n-1)//2 - m}")
```

## 3. Formulación del Modelo ILP

El Maximum Clique Problem se formula como:

$$\max \quad z = \sum_{i \in V} x_i$$

$$\text{s.a.} \quad x_i + x_j \leq 1, \quad \forall (i,j) \in \bar{E}$$

$$x_i \in \{0, 1\}, \quad \forall i \in V$$

Donde $\bar{E}$ es el conjunto de no-aristas (aristas del complemento).

```python
# Construir el conjunto de no-aristas (restricciones del modelo)
complement_edges = []
nodes = sorted(G.nodes())
node_to_idx = {node: i for i, node in enumerate(nodes)}

for i in range(len(nodes)):
    for j in range(i + 1, len(nodes)):
        if not G.has_edge(nodes[i], nodes[j]):
            complement_edges.append((i, j))

print(f"Número de restricciones de adyacencia: {len(complement_edges)}")
print(f"Variables de decisión: {n}")
```

## 4. Implementación Branch & Bound con Relajación LP

El algoritmo resuelve la relajación lineal en cada nodo del árbol de búsqueda.  
Se utiliza `scipy.optimize.linprog` para resolver el LP en cada subproblema.

```python
class BranchAndBound:
    """Branch & Bound con relajación LP para el Maximum Clique Problem."""
    
    def __init__(self, n_vars, complement_edges):
        self.n = n_vars
        self.complement_edges = complement_edges
        
        # Función objetivo: max sum(x_i) => min -sum(x_i)
        self.c = -np.ones(self.n)
        
        # Restricciones: x_i + x_j <= 1 para (i,j) en complemento
        self.A_ub = np.zeros((len(complement_edges), self.n))
        self.b_ub = np.ones(len(complement_edges))
        for k, (i, j) in enumerate(complement_edges):
            self.A_ub[k, i] = 1
            self.A_ub[k, j] = 1
        
        # Estadísticas
        self.nodes_explored = 0
        self.nodes_pruned_bound = 0
        self.nodes_pruned_infeasible = 0
        self.lp_solves = 0
        self.best_obj = 0
        self.best_solution = None
        self.history = []  # (nodos explorados, mejor cota inferior)
    
    def solve_lp(self, fixed_zeros, fixed_ones):
        """Resuelve la relajación LP con variables fijadas."""
        bounds = [(0, 1)] * self.n
        for i in fixed_zeros:
            bounds[i] = (0, 0)
        for i in fixed_ones:
            bounds[i] = (1, 1)
        
        result = linprog(
            self.c, 
            A_ub=self.A_ub, 
            b_ub=self.b_ub, 
            bounds=bounds, 
            method='highs'
        )
        self.lp_solves += 1
        return result
    
    def select_branching_variable(self, x):
        """Selecciona la variable fraccionaria más cercana a 0.5."""
        fractional = [(i, abs(x[i] - 0.5)) for i in range(self.n) 
                      if x[i] > 1e-6 and x[i] < 1 - 1e-6]
        if not fractional:
            return None
        return min(fractional, key=lambda t: t[1])[0]
    
    def is_integer(self, x, tol=1e-6):
        """Verifica si la solución es entera."""
        return all(xi < tol or xi > 1 - tol for xi in x)
    
    def solve(self, time_limit=300):
        """Ejecuta Branch & Bound. Retorna el tamaño del clique máximo."""
        start_time = time.time()
        
        # Cola de subproblemas: (fixed_zeros, fixed_ones)
        queue = deque()
        queue.append((set(), set()))
        
        while queue:
            # Verificar límite de tiempo
            elapsed = time.time() - start_time
            if elapsed > time_limit:
                print(f"\nTiempo límite alcanzado ({time_limit}s)")
                break
            
            fixed_zeros, fixed_ones = queue.pop()  # DFS (LIFO)
            self.nodes_explored += 1
            
            # Resolver relajación LP
            result = self.solve_lp(fixed_zeros, fixed_ones)
            
            # Poda por infactibilidad
            if not result.success:
                self.nodes_pruned_infeasible += 1
                continue
            
            # Valor de la relajación (cota superior del subproblema)
            ub = -result.fun
            
            # Poda por cota
            if ub <= self.best_obj:
                self.nodes_pruned_bound += 1
                continue
            
            x = result.x
            
            # Solución entera encontrada
            if self.is_integer(x):
                obj_val = sum(1 for xi in x if xi > 0.5)
                if obj_val > self.best_obj:
                    self.best_obj = obj_val
                    self.best_solution = x.copy()
                    self.history.append((self.nodes_explored, self.best_obj))
                continue
            
            # Ramificar
            branch_var = self.select_branching_variable(x)
            if branch_var is None:
                continue
            
            # Crear subproblemas: x_k = 0 y x_k = 1
            new_zeros = fixed_zeros | {branch_var}
            new_ones = fixed_ones | {branch_var}
            
            queue.append((new_zeros, fixed_ones.copy()))
            queue.append((fixed_zeros.copy(), new_ones))
            
            # Registrar progreso cada 500 nodos
            if self.nodes_explored % 500 == 0:
                self.history.append((self.nodes_explored, self.best_obj))
        
        self.elapsed_time = time.time() - start_time
        self.history.append((self.nodes_explored, self.best_obj))
        return self.best_obj, self.best_solution
```

## 5. Ejecución del Algoritmo

```python
print("Ejecutando Branch & Bound con relajación LP...")
print("="*50)

bb = BranchAndBound(n, complement_edges)
clique_size, solution = bb.solve(time_limit=300)

print(f"\n{'='*50}")
print(f"RESULTADOS")
print(f"{'='*50}")
print(f"Tamaño del clique máximo encontrado: {clique_size}")
print(f"Óptimo conocido: 34")
print(f"Gap de optimalidad: {((34 - clique_size) / 34) * 100:.1f}%")
print(f"Tiempo de ejecución: {bb.elapsed_time:.2f} s")
print(f"Nodos explorados: {bb.nodes_explored}")
print(f"LPs resueltos: {bb.lp_solves}")
print(f"Podas por cota: {bb.nodes_pruned_bound}")
print(f"Podas por infactibilidad: {bb.nodes_pruned_infeasible}")
```

```python
# Identificar los vértices del clique encontrado
if solution is not None:
    clique_nodes = [nodes[i] for i in range(n) if solution[i] > 0.5]
    print(f"Vértices del clique: {sorted(clique_nodes)}")
    
    # Verificar que es un clique válido
    subgraph = G.subgraph(clique_nodes)
    expected_edges = len(clique_nodes) * (len(clique_nodes) - 1) // 2
    actual_edges = subgraph.number_of_edges()
    print(f"\nVerificación: {actual_edges}/{expected_edges} aristas")
    print(f"¿Es clique válido? {'Sí' if actual_edges == expected_edges else 'No'}")
```

## 6. Tabla de Resultados

```python
# Resolver también la relajación LP del problema raíz para mostrar la brecha de integralidad
root_lp = bb.solve_lp(set(), set())
root_ub = -root_lp.fun

results_data = {
    'Métrica': [
        'Instancia',
        'Nodos del grafo (|V|)',
        'Aristas del grafo (|E|)',
        'Densidad del grafo',
        'No-aristas (restricciones)',
        'Cota superior LP (raíz)',
        'Óptimo conocido ω(G)',
        'Solución encontrada',
        'Gap de integralidad LP (%)',
        'Tiempo de ejecución (s)',
        'Nodos B&B explorados',
        'Relajaciones LP resueltas',
        'Podas por cota',
        'Podas por infactibilidad',
        'Eficiencia de poda (%)'
    ],
    'Valor': [
        'C125.9',
        n,
        m,
        f'{density:.4f}',
        len(complement_edges),
        f'{root_ub:.2f}',
        34,
        clique_size,
        f'{((root_ub - 34) / 34) * 100:.1f}',
        f'{bb.elapsed_time:.2f}',
        bb.nodes_explored,
        bb.lp_solves,
        bb.nodes_pruned_bound,
        bb.nodes_pruned_infeasible,
        f'{((bb.nodes_pruned_bound + bb.nodes_pruned_infeasible) / bb.nodes_explored) * 100:.1f}'
    ]
}

df_results = pd.DataFrame(results_data)
print(df_results.to_string(index=False))
```

## 7. Gráficos

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# --- Gráfico 1: Convergencia del B&B ---
ax1 = axes[0]
if bb.history:
    hist_nodes = [h[0] for h in bb.history]
    hist_obj = [h[1] for h in bb.history]
    ax1.plot(hist_nodes, hist_obj, 'b-o', markersize=4, linewidth=1.5, label='Mejor solución entera')
    ax1.axhline(y=34, color='r', linestyle='--', linewidth=1.5, label='Óptimo conocido (ω=34)')
    ax1.axhline(y=root_ub, color='g', linestyle=':', linewidth=1.5, label=f'Cota LP raíz ({root_ub:.1f})')
    ax1.set_xlabel('Nodos explorados')
    ax1.set_ylabel('Tamaño del clique')
    ax1.set_title('Convergencia del Branch & Bound')
    ax1.legend(loc='lower right')
    ax1.grid(True, alpha=0.3)

# --- Gráfico 2: Distribución de operaciones del B&B ---
ax2 = axes[1]
labels = ['Soluciones\nenteras', 'Podas\npor cota', 'Podas por\ninfactibilidad', 'Ramificaciones']
n_integer = len([h for h in bb.history if h[1] > 0])  # aproximación
n_branches = bb.nodes_explored - bb.nodes_pruned_bound - bb.nodes_pruned_infeasible
sizes = [max(1, n_integer), bb.nodes_pruned_bound, bb.nodes_pruned_infeasible, max(1, n_branches)]
colors = ['#2ecc71', '#3498db', '#e74c3c', '#f39c12']
ax2.bar(labels, sizes, color=colors, edgecolor='black', linewidth=0.5)
ax2.set_ylabel('Cantidad de nodos')
ax2.set_title('Distribución de operaciones en el árbol B&B')
ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('resultados_bb.png', dpi=150, bbox_inches='tight')
plt.show()
print("Figura guardada: resultados_bb.png")
```

```python
# --- Gráfico 3: Visualización del grafo con el clique resaltado ---
fig, ax = plt.subplots(1, 1, figsize=(10, 10))

if solution is not None:
    clique_set = set(clique_nodes)
    non_clique_nodes = [v for v in G.nodes() if v not in clique_set]
    
    pos = nx.spring_layout(G, seed=42, k=0.3)
    
    # Dibujar aristas (solo del clique para claridad)
    clique_edges = [(u, v) for u, v in G.edges() if u in clique_set and v in clique_set]
    other_edges = [(u, v) for u, v in G.edges() if u not in clique_set or v not in clique_set]
    
    nx.draw_networkx_edges(G, pos, edgelist=other_edges, alpha=0.03, 
                           edge_color='gray', ax=ax)
    nx.draw_networkx_edges(G, pos, edgelist=clique_edges, alpha=0.6, 
                           edge_color='red', width=1.5, ax=ax)
    
    # Dibujar nodos
    nx.draw_networkx_nodes(G, pos, nodelist=non_clique_nodes, 
                           node_color='lightblue', node_size=50, alpha=0.5, ax=ax)
    nx.draw_networkx_nodes(G, pos, nodelist=clique_nodes, 
                           node_color='red', node_size=120, alpha=0.9, ax=ax)
    
    ax.set_title(f'Grafo C125.9 — Clique máximo resaltado (ω = {clique_size})', fontsize=14)
    ax.axis('off')
    
    plt.tight_layout()
    plt.savefig('grafo_clique.png', dpi=150, bbox_inches='tight')
    plt.show()
    print("Figura guardada: grafo_clique.png")
```

## 8. Análisis de la Relajación LP

Analizamos la brecha de integralidad (*integrality gap*) que es la diferencia entre la cota LP y el óptimo entero.

```python
# Análisis de la solución fraccionaria de la relajación raíz
root_solution = root_lp.x

fig, ax = plt.subplots(1, 1, figsize=(12, 4))

sorted_vals = np.sort(root_solution)[::-1]
colors_bar = ['red' if v > 0.99 else ('orange' if v > 0.01 else 'lightblue') for v in sorted_vals]

ax.bar(range(n), sorted_vals, color=colors_bar, edgecolor='none', width=1.0)
ax.axhline(y=1.0, color='black', linestyle='-', linewidth=0.5)
ax.axhline(y=0.5, color='gray', linestyle='--', linewidth=0.5, alpha=0.5)
ax.set_xlabel('Variables (ordenadas por valor decreciente)')
ax.set_ylabel('Valor $x_i$ en relajación LP')
ax.set_title('Distribución de valores fraccionarios en la relajación LP raíz')
ax.set_ylim(-0.05, 1.1)

n_integer_vars = sum(1 for x in root_solution if x > 0.99 or x < 0.01)
n_fractional = n - n_integer_vars
ax.text(0.98, 0.95, f'Variables enteras: {n_integer_vars}\nVariables fraccionarias: {n_fractional}',
        transform=ax.transAxes, ha='right', va='top', fontsize=10,
        bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.tight_layout()
plt.savefig('relajacion_lp.png', dpi=150, bbox_inches='tight')
plt.show()
print("Figura guardada: relajacion_lp.png")

print(f"\nAnálisis de la relajación LP raíz:")
print(f"  Cota superior LP: {root_ub:.2f}")
print(f"  Óptimo entero conocido: 34")
print(f"  Brecha de integralidad: {root_ub - 34:.2f} ({((root_ub - 34)/34)*100:.1f}%)")
print(f"  Variables con valor 1: {sum(1 for x in root_solution if x > 0.99)}")
print(f"  Variables fraccionarias: {n_fractional}")
print(f"  Variables con valor 0: {sum(1 for x in root_solution if x < 0.01)}")
```

## 9. Resumen y Conclusiones del Experimento

### Observaciones principales:

1. **Brecha de integralidad:** La relajación LP proporciona una cota superior significativamente mayor que el óptimo entero, lo cual es una característica conocida del MCP en grafos densos.

2. **Eficacia de la poda:** A pesar de la brecha, la poda por cota permite descartar una fracción significativa de subproblemas sin necesidad de explorarlos completamente.

3. **Complejidad computacional:** El número de nodos explorados y el tiempo de ejecución evidencian la naturaleza NP-dura del problema.

4. **Validación:** La solución encontrada corresponde a un clique válido verificable en el grafo original.
